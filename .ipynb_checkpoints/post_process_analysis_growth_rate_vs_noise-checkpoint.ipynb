### Study parameters

```python
f_rev = 11.2455e3  # revolution frequency [Hz]
# Relativistic parameters
gamma_rel = 7460.52252807237 # from madx twiss
beta_rel = sqrt(1-(1/gamma_rel)**2)
# Twiss parameters
twiss = DottedDict()
twiss.beta_x = 115.75 #at IP3 #obtained from madx twiss. should be crossed check
twiss.beta_y = 1. 
twiss.alpha_x = 0 
twiss.alpha_y = 0 
twiss.gamma_x = (1+twiss.alpha_x**2)/twiss.beta_x
twiss.gamma_y = (1+twiss.alpha_y**2)/twiss.beta_y
ex_norm = 2.5e-6
ex_geom = ex_norm/(beta_rel*gamma_rel)
#ex_geom = 0.047#*1e-9 #for SCC
sigma_x = sqrt(twiss.beta_x*ex_geom)
sigma_px = sqrt(ex_geom/twiss.beta_x)
print(sigma_x, sigma_px)
particles = 15000
turns = 1000
```

### Parameters of interest, in beam sigma px

```python
Delta_values = [0.01, 0.05, 0.1, 0.5, 1.0] ## BB parameter
ksi = 0.005
gain = 0.2
A = 3.5 
```

### Calculate the emittance values with time for each one of the studies

```python
ex_dictionary = {}

for Delta_value in Delta_values:
    df = pd.read_pickle('./turn_by_turn_data/Qx0.78_noise{}_BB0.005_feedback0.2_betax115.75_H_particles15000_turns1000.pkl'.format(Delta_value, ksi))
    ex_dictionary['Noise'+str(Delta_value)+ '_BB'+str(ksi)] = [] # the name of the list on the dicitonary indicates the values of the BB parameter and of the strength noise
    for turn in range(turns):
        ex_dictionary['Noise'+str(Delta_value)+ '_BB'+str(ksi)].append(cmp_emit(df.at[turn,'x']/sqrt(twiss.beta_x), df.at[turn,'px']*sqrt(twiss.beta_x) ))
```

### Apply linear fit at the emitance values of each study.
During the first turns a flactuation of the emittance values is observed due to mismatched beam. Therfore the first 200 turns are excluded from the fit

```python
fig, ax = plt.subplots()

turns_list = np.arange(0,turns)

for Delta_value in Delta_values:

    current_study = ex_dictionary['Noise'+str(Delta_value)+ '_BB'+str(ksi)]
    [m_h,b_h], cov_h = np.polyfit(turns_list, current_study/current_study[0], 1, cov= True)
    ax.scatter(Delta_value, m_h, s = 40)
    
    ax.set_yscale('log')
    ax.set_xscale('log')
    ax.set_xlabel('Noise strength')
    ax.set_ylabel(r'$\frac{1}{\epsilon_0} \frac{d\epsilon}{dn}$')
    
plt.legend()
```

### Compare emittance growth rate with the expected rate

```python
fig, ax = plt.subplots()

turns_list = np.arange(0,turns)

theory_rate = []
errors = []
growth_rate = []

for Delta_value in Delta_values:

    current_study = ex_dictionary['Noise'+str(Delta_value)+ '_BB'+str(ksi)]
    [m_h,b_h], cov_h = np.polyfit(turns_list, current_study/current_study[0], 1, cov= True)
    err_h = np.sqrt(np.diag(cov_h))
    
    #errors.append(err_h)
    #growth_rate.append(m_h)
    theory_rate.append(A*((ksi**2)*((Delta_value**2))/(2*gain**2)))
    
    
    ax.scatter(Delta_value, m_h, s = 40,c = 'k')
    ax.errorbar(Delta_value, m_h, yerr=[err_h[0]], fmt= 'o')
    
ax.set_ylim(10e-8, 10e-3)
ax.set_yscale('log')
ax.set_xscale('log')
ax.set_xlabel('Noise strength')
ax.set_ylabel(r'$\frac{1}{\epsilon_0} \frac{d\epsilon}{dn}$', fontsize = 25)
    
ax.plot(Delta_values, theory_rate, c = 'k')   

plt.tight_layout()
plt.savefig('./figures/emittance_growth_rate_Qx0.78_noisescan_BB0.005_feedback0.2_betax115.75_H_particles15000_turns1000.png')
```
